```python
from PIL import Image
import numpy as np

def gif_to_rgb_bitstring(gif_path):
    gif = Image.open(gif_path)
    
    all_frames_bitstrings = []
    
    for frame in range(gif.n_frames):
        gif.seek(frame)
        
        rgb_frame = gif.convert('RGB')
        rgb_array = np.array(rgb_frame)
        flat_rgb_array = rgb_array.flatten()
        bitstring = ''.join(f'{pixel:08b}' for pixel in flat_rgb_array)
        all_frames_bitstrings.append(bitstring)
    
    return all_frames_bitstrings

gif_path = 'multicolor.gif'
rgb_bitstrings = gif_to_rgb_bitstring(gif_path)
print(f"First frame RGB bitstring (first 256 bits): {rgb_bitstrings[0]}")
```
